```python
def calculate_grade(marks):
    if marks >= 90:
        return "A+"
    elif marks >= 80:
        return "A"
    elif marks >= 70:
        return "B"
    elif marks >= 60:
        return "C"
    elif marks >= 50:
        return "D"
    else:
        return "F"

# Function to display all student grades
def display_grades(students):
    print("\n--- Student Grades ---")
    for name, marks in students.items():
        grade = calculate_grade(marks)
        print(f"{name} scored {marks} and got grade: {grade}")


students = {}  # empty dictionary to store data

while True:
    name = input("\nEnter student name (or type 'exit' to stop): ")
    if name.lower() == 'exit':
        break

    marks = int(input(f"Enter marks for {name}: "))
    students[name] = marks  # store data in dictionary


display_grades(students)

print("\nProgram finished. Thank you!")
```
